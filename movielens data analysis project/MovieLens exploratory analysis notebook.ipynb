```python
#Visualizing how popularity of genres has changed over the past years
```

```python
df_movie_genre_year=movie.drop(['movie id','movie title','release date'],axis=1)
df_movie_genre_year
```

```python
dfff=df_movie_genre_year.set_index('release year').groupby('release year').sum()
dfff
```

```python
movie_2=movie[['movie title','release year']].groupby('release year').count()
movie_2
```

```python
df3=dfff.merge(movie_2,on='release year').rename(columns={'movie title':'total number of movies released'})
df3
```

```python
df4=df3.div(df3['total number of movies released'], axis= 0) * 100
del df4['total number of movies released']
df4
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(30,15))
sns.heatmap(df4.T)
```

**Observation:**
 - Drama was the most released type of movies throughout the years.Documentary and children were the least released
 - Children movies had good release contributions from 1960 to 1980 but in recent years that number has dropped.

```python
df9=movie.merge(rating,on='movie id')[['user id','movie title','rating']]
df9
```

```python
dff11=df10[df10['count']>100].sort_values(by='mean',ascending=False)
dff11[['movie title']].reset_index(drop=True).head(25)
```

```python
#Checking if 
#1-Men watch more drama than women
#2-Women watch more Sci-Fi than men
#3-Men watch more Romance than women
```

```python
df_final=user.merge(rating,on='user id').merge(movie,on='movie id')
df_final
```

```python
df_gender=df_final.groupby('gender').count()['rating'].reset_index()
df_gender
```

```python
df_drama=df_final[df_final['Drama']==1][['gender','Drama']].groupby('gender').sum()
df_drama
```

```python
df_romance=df_final[df_final['Romance']==1][['gender','Romance']].groupby('gender').sum()
df_romance
```

```python
df_scifi=df_final[df_final['Sci-Fi']==1][['gender','Sci-Fi']].groupby('gender').sum()
df_scifi
```

```python
df_result=df_gender.merge(df_drama,on='gender').merge(df_romance,on='gender').merge(df_scifi,on='gender')
df_result=df_result.rename(columns={'rating':'total rating received'})
df_result
```

```python
df_result=df_result.set_index('gender')
df_result=df_result.div(df_result['total rating received'], axis= 0) * 100
df_result
```

```python
del df_result['total rating received']
df_result
```

**Final conlusion:**
 - Men watch more drama than women:False statement
 - Women watch more Sci-Fi than men: False statement
 - Men watch more Romance than women: False statement
